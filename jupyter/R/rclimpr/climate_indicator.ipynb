<p style='text-align: justify; font-size: 16px;'>
This notebook illustrates the integration of climate data from different data sources, such as CPTEC/INPE, Era5-Land, among others, into the Earth Observation Data Cubes tuned for Health Response platform (EODCtHRS). This is an instance of the Brazil Data Cube (BDC) project, which has the capability to accommodate large volumes of geospatial data and offers a variety of services and applications for analysis and visualization. This example will demonstrate how these climate data can be integrated into EODCtHRS to generate climate indicators, enabling a more comprehensive analysis of environmental impacts on public health in Brazil. The first collection of climate data is generated based on data produced by <a href="https://rfsaldanha.github.io/data-projects/era5land-daily-latin-america.html">Raphael Saldanha</a>. He created daily aggregates from some ERA5-Land indicators for the Latin America region.
</p>

<p style='text-align: justify; font-size: 16px;'>
    <a href="#climate_workflow">Figure 1</a> explains this data process workflow.    
</p>

<figure style="text-align: center;">
  <img id="climate_workflow" src=" " />
  <figcaption style='text-align: center;'><b>Figure 1</b> - Climate Data Production Flow. Source: adapted from <a href="#ref_saldanha_2023">Saldanha <i>et.al</i>, 2023</a>.</figcaption>
</figure>


<p style='text-align: justify; font-size: 16px;'>
The <a href="https://github.com/Harmonize-Brazil/rclimpr">EODCtHRS <b>R</b> <b>Clim</b>ate <b>Pr</b>ocessing Package (rclimpr)</a> is composed of a set of functions for reading files in NetCDF format, that enable spatial aggregations, such as grouping by municipalities, and temporal aggregations, such as epidemiological weeks and months. As output, it generates files in raster formats, Cloud-Optimized GeoTIFF (COG), and vector formats, such as GeoJSON and Shapefile, aiming to provide users with the most suitable data format for their specific analysis and visualization. As input data, daily aggregations of temperature and precipitation indicators from Era5-Land, produced by Raphael Saldanha, were utilized. <a href="#prococol_rclimpr">Figure 2</a> presents the data process workflow.    
</p>

<figure style="text-align: center;">
  <img id="prococol_rclimpr" src=" " style="width: 200px";/>
  <figcaption style='text-align: center;'><b>Figure 2</b> - Protocol climate data in rclimpr package.</figcaption>
</figure>

<p style='text-align: justify; font-size: 16px;'>
<b>NOTE:</b> This R package, in its initial version, is optimized for handling a specific data format, NetCDF with one variable, automatically converting values from Kelvin to Celsius (temperature) and from meters to millimeters (precipitation) based on user-defined input parameters. Future updates will include additional enhancements and user-defined options.
</p>

#### <span style="color:#336699">Complete process</span>
<hr style="border:1px solid #0077b9;">

<p style='text-align: justify; font-size: 16px;'>
A complete process is available in <a href="https://github.com/Harmonize-Brazil/edpu"> <b>E</b>ODCtHRS <b>D</b>ata <b>PU</b>blisher (EDPU)</a> package. It is a package developed by Harmonize Brazil Team - INPE in Python to provides a set of scripts to process the health indicator data produced by LIS to be accommodated in the EODCtHRS component. Although the dataset of Harmonize project contains heterogeneous data, such as tabular, vector and raster data, the workflow to accommodate these data are the same. All of them are published into GeoServer as a layer and their metadata are published into STAC. So, taking it into consideration, we developed this package to optimize the flow of publishing the data into EODCtHRS. This workflow is represented in the <a href="#general_workflow">Figure 3</a> below.
